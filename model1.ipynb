### **My first ML project**

## **Loading data **

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

## Data preparation.

### spliting data into x and y

```python
#displaying the y-axis.
Y=df['logS']
Y
```

```python
#droping y axis values (logS column) and displaying x axis values.
x = df.drop('logS', axis=1)
x
```

### Split dataset into training set and test set.

```python
#import spliting library from sklearn.model_selection.

from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state = 100)

#look at x_train dataset. this will have 80% of the total data.
x_train
```

```python
#look at x_test dataset.
#test set will have 20% of the total data.
x_test
```

## Model Building.

### Linear Regression

### **Training the model.**

```python
#import linearregression.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, Y_train)
```

### Applying the model.

```python
Y_lr_train_pred = lr.predict(x_train)
Y_lr_test_pred = lr.predict(x_test)

Y_lr_train_pred
```

```python
Y_lr_test_pred
```

## Evaluating the model performance.


```python
Y_train
```

```python
Y_lr_train_pred
```

```python
#evaluating the model.
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(Y_train, Y_lr_train_pred)
lr_train_r2 = r2_score(Y_train, Y_lr_train_pred)

lr_test_mse = mean_squared_error(Y_test, Y_lr_test_pred)
lr_test_r2 = r2_score(Y_test, Y_lr_test_pred)

lr_train_mse

print('LR MSE(Train):', lr_train_mse)
print('LR r2(Train):',lr_train_r2)
print('LR MSE(Test):',lr_test_mse)
print('LR R2(Test):',lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()

lr_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']

lr_results
```

## **Random Forest Model**

### Training the model.

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth =2, random_state=100)
rf.fit(x_train, Y_train)
```

### Applying the model.


```python
Y_rf_train_pred = rf.predict(x_train)
Y_rf_test_pred = rf.predict(x_test)
```

### Evaluating the model.

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(Y_train, Y_lr_train_pred)
rf_train_r2 = r2_score(Y_train, Y_lr_train_pred)

rf_test_mse = mean_squared_error(Y_test, Y_rf_test_pred)
rf_test_r2 = r2_score(Y_test, Y_rf_test_pred)

rf_train_mse

print('RF MSE(Train):', rf_train_mse)
print(['RF R2(Train):', rf_train_r2])
print(['RF MSE(Test):', rf_test_mse])
print(['RF R2(Test):', rf_test_r2])
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

# Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
#reseting the indexing of the raws.
df_models.reset_index(drop= True)
```
